```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
start_time = datetime(2026, 6, 10, 20, 30)
```

```python
q1_pool = [
    "Let's go Knicks!! MSG is rocking tonight!", "Spurs hitting everything early, stay calm.",
    "Wemby is a problem in the paint.", "Bad turnover there, wake up Brunson!",
    "Our shooting looks ice cold to start.", "Down 8 early, need defensive stops.",
    "Brutal first quarter. We can't buy a bucket.", "Terrible shot selection right now."
]

q2_pool = [
    "Down 14... this is getting incredibly ugly.", "Are we really laying an egg at home in the Finals?",
    "Absolute disaster class on offense.", "Can anyone please hit a wide-open three?!",
    "This is embarrassing to watch on national TV.", "Down 22 points at the half. I am sick.",
    "No energy, no defense, no adjustments. Disastrous.", "Halftime can't come soon enough. Total failure."
]

q3_pool = [
    "Down 25. This game is officially over.", "WE ARE DOWN 29 POINTS. THIS IS TRAGIC.",
    "Wait, nice hustle play by Josh Hart there.", "Brunson with a tough and-one. Keep fighting!",
    "A quick 6-0 run? Okay, don't give me hope.", "Deficit down to 19! MSG is waking up!",
    "OG Anunoby with a massive block!! Let's goooo!", "Holding them to 14 points this quarter is huge.",
    "Down 10 heading into the 4th. Anything is possible now!"
]

q4_pool = [
    "BRUNSON THREE!!! Single digits baby!!", "Defense is absolute clamps right now!!",
    "TIE GAME!!! MY VOICE IS COMPLETELY GONE!!!", "Oh no, Wemby hits a tough midrange. Down 3.",
    "BRUNSON FLOATER!! WE HAVE THE LEAD!!!", "12 seconds left, down 1. My heart is pounding.",
    "OG ANUNOBY PUTBACK AT THE BUZZER!!!! WE WON!!!!!", "GREATEST COMEBACK IN NEW YORK KNICKS HISTORY!!!",
    "I AM LITERALY CRYING RIGHT NOW WHAT A GAME", "NEW YORK STAND UP!! PLAY THE MANNIX!!"
]
```

```python
simulated_data = []
```

```python
simulated_data = []

for _ in range(500):
   
    minutes_elapsed = random.randint(0, 180)
    post_time = start_time + timedelta(minutes=minutes_elapsed)
    
   
    if minutes_elapsed <= 45:
        comment = random.choice(q1_pool)
        score_diff = random.randint(-15, -2)
    elif minutes_elapsed <= 90:
        comment = random.choice(q2_pool)
        score_diff = random.randint(-26, -14)
    elif minutes_elapsed <= 135:
        comment = random.choice(q3_pool)
        score_diff = -29 if 100 <= minutes_elapsed <= 110 else random.randint(-28, -10)
    else:
        comment = random.choice(q4_pool)
        score_diff = random.randint(-9, 2)
        
        if minutes_elapsed >= 175:
            score_diff = 1 
            comment = "OG ANUNOBY PUTBACK AT THE BUZZER!!!! WE WON!!!!!"

  
    vs = analyzer.polarity_scores(comment)
    
    simulated_data.append({
        'Timestamp': post_time,
        'Score_Differential': score_diff,
        'Fan_Commentary': comment,
        'Sentiment_Score': vs['compound']
    })
```

```python
df_simulated = pd.DataFrame(simulated_data)
df_simulated = df_simulated.sort_values(by='Timestamp')
```

```python
df_simulated.to_csv('knicks_simulated_stream.csv', index=False)
print("Success! Generated 'knicks_simulated_stream.csv' with 500 rows of continuous stream data.")
```
